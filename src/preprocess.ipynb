## Note:
removed users who have not completed the course or had no remediations

## Coding details:
sex: {sex_f, sex_m} | if both 0, not given <br/>
age: {-1: not given, 0: 18-24, 1: 25-34, 2: 35-44, 3: 45-54, 4: 55-64, 5: 65 or older} <br/>
education: {-1: not given, 0: high school or some collage, 1: Associate degree (according to wiki), 2: Collage or some masters, 3: Masters or Doctorate} <br/>
coin_familiarity: {-1: not given, 0: 1, 1: 5} | There were no other values in between <br/>
coin_interest: {-1: not given, 0: 1, 1: 5} | There were no other values in between <br/>


```python
import pandas as pd
import os
import collections

total_count = 0
completed_count = 0
have_remediations = 0

e = collections.Counter()
df_user = pd.DataFrame(columns=['username', 'sex_f', 'sex_m', 'age', 'education', 'coin_familiarity', 'coin_interest',
                               'pretest', 'posttest', 'nlg'])
for file in os.listdir('../data/data(COIN)/'):
    if '.csv' in file:
        batch = file.split("Batch ")[1].split(" Course")[0]
        course = file.split(" Course ")[1].split(" - ")[0]
        user_id = file.split("USER ID ")[1].split('.csv')[0]
        username = batch + '-' + course + '-' + user_id
        
        total_count += 1
        
        d = pd.read_csv('../data/data(COIN)/'+file, encoding='iso8859', header=None, sep='\n')

        if d[0].str.split(',')[0][3]=='no':
#             print("NOT COMPLETED", username)
            continue
        completed_count += 1
        
        found = False
        col_type = d[0].str.split(',')
        for l in range(len(col_type)):
            if col_type[l][0]=='remediation':
                found = True
                break
        if found is False:
#             print("NO REM", username)
            continue
        have_remediations += 1
        
        # DEMOGRAPHICS
        sex_f = 0
        sex_m = 0
        sex = d[0].str.split(',')[1][2]
        if sex == 'Male':
            sex_m = 1
        elif sex == 'Female':
            sex_f = 1
        

        age = d[0].str.split(',')[1][3]
        if age == '18-24':
            age = 0
        elif age == '25-34':
            age = 1
        elif age == '35-44':
            age = 2
        elif age == '45-54':
            age = 3
        elif age == '55-64':
            age = 4
        elif age == '65 or older':
            age = 5
        elif age == '':
            age = -1
        else:
            print("age UNKNOWN")


        education = d[0].str.split(',')[1][4]
        if "High School" in education or "Some College" in education:
            education = 0
        elif "Associate Degree" in education:
            education = 1
        elif "College" in education or "Some Masters" in education:
            education = 2
        elif "Masters" in education or "Doctorate" in education:
            education = 3
        elif education == "":
            education = -1
        else:
            print("education UNKNOWN", education)


        coin_familiarity = d[0].str.split(',')[1][5]
        if coin_familiarity == '1':
            coin_familiarity = 0
        elif coin_familiarity == '5':
            coin_familiarity = 1
        elif coin_familiarity == '':
            coin_familiarity = -1
        else:
            print("coin_familiarity UNKNOWN")

        coin_interest = d[0].str.split(',')[1][6]
        if coin_interest == '1':
            coin_interest = 0
        elif coin_interest == '5':
            coin_interest = 1
        elif coin_interest == '':
            coin_interest = -1
        else:
            print("coin_interest UNKNOWN")

        
        pretest = d[0].str.split(',')[1][23]
        
        
        pretest = int(pretest)
        
        posttest = ''
        col_type = d[0].str.split(',')
        for l in range(len(col_type)):
            if col_type[l][0]=='posttest':
                posttest = col_type[l][14]
                break
        if posttest=='':
            print("NO POSTTEST FOUND", username)
        posttest = int(posttest)
        
        
        if posttest>pretest:
            nlg = round((posttest-pretest)/(12-pretest)*100, 2)
        elif posttest==pretest:
            nlg = 0.0
        else:
            nlg = round((posttest-pretest)/pretest*100, 2)
            
        
        
        
        df_user = df_user.append({'username':username, 'sex_f': sex_f, 'sex_m':sex_m, 'age':age, 'education':education, 
                                  'coin_familiarity': coin_familiarity, 'coin_interest': coin_interest, 
                                  'pretest': pretest, 'posttest': posttest, 'nlg': nlg}, ignore_index=True)


        # Video
        c=0
        video_data = d[0].str.split(',')
        for l in range(len(video_data)):
            if video_data[l][0]=='video':
                c+=1
        if c!=12:
            print("SOME VIDEOS MISSING", username)

print("Total Participants:", total_count, "| Completed:", completed_count, "| Have remediations:", have_remediations)            
```

```python
# NLG HIGH LOW
nlg_median = df_user.describe().loc['50%'].tolist()[0]
df_user.loc[df_user['nlg']<nlg_median, 'nlg_high_low'] = 0
df_user.loc[df_user['nlg']>=nlg_median, 'nlg_high_low'] = 1

df_user.to_pickle('../processed_data/user_df.pkl')
df_user
```
